```
def osszegkiiro(*vartuple):
    sum = 0
    for var in vartuple[:-1]:
        sum += var
        print(var, "+",end =" ")

    sum += vartuple[-1]
    print(vartuple[-1], '=', sum)
    return
```

```
osszegkiiro(1, 2, 3, 4)
osszegkiiro(33, 452, 0)
```
